# Problem 72
Invert a binary tree.

For example, given the following tree:

```
    a
   / \
  b   c
 / \  /
d   e f
```

should become:

```
  a
 / \
 c  b
 \  / \
  f e  d
```

---
## Solution

````python
# solution code

class Node:
    def __init__(self, val, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right


def serialize(root):
    tree_structure = []
    tree_structure.append(root.val)
    if(root.left != None):
        if(root.right == None):
            tree_structure.append(serialize(root.left))
            tree_structure.append(["None"])
        else:
            tree_structure.append(serialize(root.left))
    if(root.right != None):
        if(root.left == None):
            tree_structure.append(["None"])
        tree_structure.append(serialize(root.right))
    return tree_structure


def invert_deserialize(serialize_structure):
    if(len(serialize_structure) > 1):
        node = Node(serialize_structure[0])
        node.left = invert_deserialize(serialize_structure[2])
        node.right = invert_deserialize(serialize_structure[1])
    else:
        node = Node(serialize_structure[0])
    return node


def reflect_tree(tree):
    serialize_tree = serialize(tree)
    tree_flipped = invert_deserialize(serialize_tree)
    return tree_flipped
````

````python
# Found function on StackOverflow which displays binary trees
# https://stackoverflow.com/questions/34012886/print-binary-tree-level-by-level-in-python
def PrintTree(root):
    def height(root):
        return 1 + max(height(root.left), height(root.right)) if root else -1  
    nlevels = height(root)
    width =  pow(2,nlevels+1)

    q=[(root,0,width,'c')]
    levels=[]

    while(q):
        node,level,x,align= q.pop(0)
        if node:            
            if len(levels)<=level:
                levels.append([])
        
            levels[level].append([node,level,x,align])
            seg= width//(pow(2,level+1))
            q.append((node.left,level+1,x-seg,'l'))
            q.append((node.right,level+1,x+seg,'r'))

    for i,l in enumerate(levels):
        pre=0
        preline=0
        linestr=''
        pstr=''
        seg= width//(pow(2,i+1))
        for n in l:
            if(str(n[0].val) == "None"):
                valstr = ''
            else:
                valstr= str(n[0].val)
            if n[3]=='r':
                linestr+=' '*(n[2]-preline-1-seg-seg//2)+ '¯'*(seg +seg//2)+'\\'
                preline = n[2] 
            if n[3]=='l':
               linestr+=' '*(n[2]-preline-1)+'/' + '¯'*(seg+seg//2)  
               preline = n[2] + seg + seg//2
            pstr+=' '*(n[2]-pre-len(valstr))+valstr #correct the potition acording to the number size
            pre = n[2]
        print(linestr)
        print(pstr)   
````

## Solution Explained

### `reflect_tree(tree)` solution
This code implements the inversion of a binary tree. The code defines a class called `Node` that represents a node in the binary tree. Each node has a value, a left child, and a right child. The code also defines three functions: `serialize,` `invert_deserialize,` and `reflect_tree.`

The `serialize` function takes a binary tree as an input and returns a list that represents the structure of the tree. The function first adds the value of the root node to the list. Then, it checks if the root has a left child. If it does, the function recursively calls itself on the left child and appends the result to the list. If the root has a right child, the function does the same thing with the right child. If the root has both a left and a right child, the function first serializes the left child and then the right child.

The `invert_deserialize` function takes the list generated by the `serialize` function as an input and returns a binary tree with the same structure. The function first creates a node with the value at the first index of the list. Then, it checks if the list has more than one element. If it does, the function recursively calls itself on the second and third elements of the list, which represent the right and left children of the node, respectively. Finally, the function sets the left and right children of the node to the results of the recursive calls.

The `reflect_tree` function takes a binary tree as an input, uses the `serialize` function to get the tree structure as a list, and then uses the `invert_deserialize` function to invert the tree. The function then returns the inverted tree.

Overall, this code implements the inversion of a binary tree in a recursive and efficient way.
