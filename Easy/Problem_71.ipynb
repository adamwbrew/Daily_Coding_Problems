# Problem 71
Given a mapping of digits to letters (as in a phone number), and a digit string, return all possible letters the number could represent. You can assume each valid number in the mapping is a single digit.

For example if `{“2”: [“a”, “b”, “c”], 3: [“d”, “e”, “f”], …}` then `“23”` should return `[“ad”, “ae”, “af”, “bd”, “be”, “bf”, “cd”, “ce”, “cf"]`.

---
## Solution

```python
# solution code

def mapping_combinations(mapping, digit):
    if not isinstance(mapping, dict):
        return f"Mapping of digits must be a dictionary. You inputted a {type(mapping)}."
    if not isinstance(digit, str):
        return f"The digit input must be a string. You inputted a {type(digit)}."
    mapped_letters = [mapping[d] for d in digit]
    combos = mapped_letters[0]
    for i in range(1, len(mapped_letters)):
        new_combos = []
        for c in combos:
            for letter in mapped_letters[i]:
                new_combos.append(c + letter)
        combos = new_combos
    return combos
```

## Solution Explained

### function solution
The function `mapping_combinations` that takes two arguments, a `mapping` dictionary and a `digit` string. The function returns all possible letter combinations that could be formed from the given digit string, based on the mapping provided in the mapping dictionary.

The function first checks if the `mapping` argument is a dictionary and the digit argument is a string. If either of these conditions are not satisfied, the function returns an error message.

Next, the function initializes a list called `mapped_letters` that contains the values of each key in the `mapping` dictionary that corresponds to the digits in the `digit` string.

The function then initializes a variable called `combos` with the first element of `mapped_letters`. It then iterates through the remaining elements of `mapped_letters` and, for each element, it generates new combinations by concatenating each element of `combos` with each element of the current element of `mapped_letters`. The resulting combinations are stored in a new list called `new_combos`.

Finally, the function updates `combos` to be equal to `new_combos`, and continues to generate new combinations until all digits in `digit` have been processed. The final value of `combos` is returned as the list of all possible letter combinations.
